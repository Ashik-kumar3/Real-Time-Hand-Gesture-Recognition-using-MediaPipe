```python
import cv2
import mediapipe as mp
```

```python
mph = mp.solutions.hands
mp_draw = mp.solutions.drawing_utils

data = mph.Hands(max_num_hands=1)

video = cv2.VideoCapture(0)

while True:
    suc, img = video.read()

    img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

    result = data.process(img1)

    tip_id = [4,8,12,16,20]
    lmlist =[]

    if result.multi_hand_landmarks:
        for x in result.multi_hand_landmarks:
            for id,lm in enumerate(x.landmark):
                # print(id,lm)
                cx = lm.x
                cy = lm.y

                lmlist.append([id,cx,cy])
                if len(lmlist)!=0 and len(lmlist)==21:
                    finger = []
                    if lmlist[12][1]<lmlist[20][1]:     #right hand
                        if lmlist[4][1]>lmlist[3][1]:   #closed
                            finger.append(0)
                        else :                          #opened
                            finger.append(1)
                    else:                               #left hand
                        if lmlist[4][1]<lmlist[3][1]:   #closed
                                finger.append(0)
                        else :                          #opened
                            finger.append(1)
                    for i in range(1,5):
                        if lmlist[tip_id[i]][2]>lmlist[tip_id[i]-2][2]:     # y value of tip will be greater than y value of tip-2
                            finger.append(0)
                        else:
                            finger.append(1)
                    # print(finger)
                    # print count

                    if finger ==[0,1,1,0,0]:
                        text = 'Peace'
                    elif finger == [1,1,0,0,1]:
                        text = 'Rock'
                    elif finger == [1,1,1,1,1]:
                        text = 'Stop'
                    elif finger == [1,0,0,0,1]:
                        text = 'Call me'
                    elif finger == [0,1,0,0,0]:
                        text = 'Out'
                    elif finger == [0,0,1,1,1]:
                        text = 'OK'
                    elif finger == [0,0,0,0,0]:
                        text = 'fist'
                    elif finger == [1,1,0,0,0]:
                        text = 'loser'
                    elif finger == [1,0,0,0,0]:
                        if lmlist[4][2]<lmlist[1][2]:
                            text = 'Thumbs Up'
                        else :
                            text ='Thumbs down'

                    else:
                        text ='Unknown'
                    # 
                    cv2.putText(img,text,(35,150),cv2.FONT_HERSHEY_PLAIN,5,(0,255,0),5)
              
            mp_draw.draw_landmarks(img,x,mph.HAND_CONNECTIONS)

    cv2.imshow('frame',img)
    if cv2.waitKey(1) & 0xff == 27:
        break
video.release()
cv2.destroyAllWindows()
```
